```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#DataFrame 1: csv
#Para gerar novos gráficos rode essa sessão de código novamente
file = 'master.csv'
df = pd.read_csv(file)
```

```python
#Número de suicídios cometidos por homens no Brasil no ano de 2000
df.set_index('age', inplace = True)
df.query('country == "Brazil" and year == ["2000"] and sex == "male"')[['suicides_no']].plot.bar()
```

```python
#Número de suicídios cometidos por homens no Brasil no ano de 2015
df.set_index('age', inplace = True)
df.query('country == "Brazil" and year == ["2015"] and sex == "male"')[['suicides_no']].plot.bar()
```

```python
#Número de suicídios cometidos por mulheres no Brasil no ano de 2000
df.set_index('age', inplace = True)
df.query('country == "Brazil" and year == ["2000"] and sex == "female"')[['suicides_no']].plot.bar()
```

```python
#Número de suicídios cometidos por mulheres no Brasil no ano de 2015
df.set_index('age', inplace = True)
df.query('country == "Brazil" and year == ["2015"] and sex == "female"')[['suicides_no']].plot.bar()
```

```python
#DataFrame 2: média de suicidios homens
#index 0 para o ano 2000
#index 1 para o ano 2015

#total mortes de homens (todas as idades)
sh2000 = [215,900,1989,1200,1037,42]
sh2015 = [470,1693,3238,1897,1389,79]
sh2000.sort()
sh2015.sort()

#total da população de homens
poph2000 = 1077633+7486883+19576126+14100257+17494250+17927425
poph2015 = 2448560+12987357+26989063+17625976+17569045+16057040

#variancia e dispersão
varh2000 = round(np.var(sh2000),2) #4.166718e+05
varh2015 = round(np.var(sh2015),2) #1.049806e+06

df3 = pd.DataFrame(np.array([[sum(sh2000),popm2000,(sum(sh2000) / poph2000)*100,(sum(sh2000) / poph2000)*100000, (sh2000[2]+sh2000[3])/2, varh2000, varh2000**0.5],
                             [sum(sh2015),popm2015,(sum(sh2015) / poph2015)*100,(sum(sh2015) / poph2015)*100000, (sh2015[2]+sh2015[3])/2, varh2015, varh2015**0.5]]), columns=['Homens','Total de homens','Média(%)','Média(/100k hab)','Mediana','Variância','Desvio padrão'])
df3
```
